**1. Basic Line Plot**

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y,label="Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.legend()
plt.show()
```

**2.Histogram**

```python
data =[7,8,5,7,10,6,5,9]
plt.hist(data,bins=5,color='blue',alpha=0.7)
plt.title("Histogram")
plt.show()
```
